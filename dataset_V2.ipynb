## Ajout des cartons

Fonction renvoyant un dataframe du cumul des cartons jaunes par journée, par équipe

```python
def yellow_made(df, covid):
    # Création du dictionnaire des équipes de la saison
    teams = {}
    for i in df['HomeTeam'].unique():
        teams[i] = []
    # Journée par journée, on ajoute le nombre de cartons jaunes
    for _, row in df.iterrows():
        yellow_home = row['HY']
        yellow_away = row['AY']
        teams[row['HomeTeam']].append(yellow_home)
        teams[row['AwayTeam']].append(yellow_away)
    # On ajoute ces données dans un dataframe équipe par équipe, journée par journée
    colonnes = list(range(1, 2*(len(teams)-1)+1))
    if covid:
        colonnes = list(range(1, 28))
    df_yellow = pd.DataFrame(data = teams, index=colonnes).T
    df_yellow.insert(0, 0, 0)
    # On cumule le nombre de cartons
    for i in colonnes:
        df_yellow[i] = df_yellow[i] + df_yellow[i-1]
    return df_yellow
```

Fonction renvoyant un dataframe du cumul des cartons rouges par journée, par équipe

```python
def red_made(df, covid):
    # Création du dictionnaire des équipes de la saison
    teams = {}
    for i in df['HomeTeam'].unique():
        teams[i] = []
    # Journée par journée, on ajoute le nombre de cartons rouges
    for _, row in df.iterrows():
        red_home = row['HR']
        red_away = row['AR']
        teams[row['HomeTeam']].append(red_home)
        teams[row['AwayTeam']].append(red_away)
    # On ajoute ces données dans un dataframe équipe par équipe, journée par journée
    colonnes = list(range(1, 2*(len(teams)-1)+1))
    if covid:
        colonnes = list(range(1, 28))
    df_red = pd.DataFrame(data = teams, index=colonnes).T
    df_red.insert(0, 0, 0)
    # On cumule le nombre de cartons
    for i in colonnes:
        df_red[i] = df_red[i] + df_red[i-1]
    return df_red
```

Fonction ajoutant les cartons par chaque équipe chaque journée

```python
def get_cards(df, covid):
    # On récupère les dataframes de cartons par journée
    yellow_ma = yellow_made(df, covid)
    red_ma = red_made(df, covid)
    
    # On calcule le nombre de matches par journée
    matches_per_day = int(len(df)/(yellow_ma.shape[1]-1))

    # On initialise les tableaux
    day = 0
    H_yellow_ma = []
    A_yellow_ma = []
    H_red_ma = []
    A_red_ma = []

    # Pour chaque match, on ajoute les cartons
    for index, row in df.iterrows():
        H_yellow_ma.append(yellow_ma[day][row['HomeTeam']])
        A_yellow_ma.append(yellow_ma[day][row['AwayTeam']])
        H_red_ma.append(red_ma[day][row['HomeTeam']])
        A_red_ma.append(red_ma[day][row['AwayTeam']])
        # On incrémente day quand on passe d'une journée à une autre
        if ((index + 1)% matches_per_day) == 0:
            day += 1
    
    df['HomeYellowlMade'] = H_yellow_ma
    df['AwayYellowMade'] = A_yellow_ma
    df['HomeRedMade'] = H_red_ma
    df['AwayRedMade'] = A_red_ma

    return df
```

```python
def conceded(df, covid):
    # Création du dictionnaire des équipes de la saison
    teams = {}
    for i in df['HomeTeam'].unique():
        teams[i] = []
    # Journée par journée, on ajoute le nombre de buts encaissés
    for _, row in df.iterrows():
        goal_home = row['FTHG']
        goal_away = row['FTAG']
        teams[row['HomeTeam']].append(goal_away)
        teams[row['AwayTeam']].append(goal_home)
    # On ajoute ces données dans un dataframe équipe par équipe, journée par journée
    colonnes = list(range(1, 2*(len(teams)-1)+1))
    if covid:
        colonnes = list(range(1, 28))
    df_conc = pd.DataFrame(data = teams, index=colonnes).T
    df_conc.insert(0, 0, 0)
    # On cumule le nombre de buts encaissés
    for i in colonnes:
        df_conc[i] = df_conc[i] + df_conc[i-1]
    return df_conc
```

Fonction ajoutant les buts marqués et encaissés par chaque équipe chaque journée

```python
def get_goals(df, covid):
    # On récupère les dataframes de buts marqués et encaissés par journée
    goal_sc = scored(df, covid)
    goal_co = conceded(df, covid)
    
    # On calcule le nombre de matches par journée
    matches_per_day = int(len(df)/(goal_sc.shape[1]-1))

    # On initialise les tableaux
    day = 0
    M_week = []
    H_scored = []
    H_conceded = []
    A_scored = []
    A_conceded = []
    H_diff = []
    A_diff = []

    # Pour chaque match, on ajoute les buts
    for index, row in df.iterrows():
        M_week.append(day+1)
        H_scored.append(goal_sc[day][row['HomeTeam']])
        H_conceded.append(goal_co[day][row['HomeTeam']])
        A_scored.append(goal_sc[day][row['AwayTeam']])
        A_conceded.append(goal_co[day][row['AwayTeam']])
        H_diff.append(goal_sc[day][row['HomeTeam']] - goal_co[day][row['HomeTeam']])
        A_diff.append(goal_sc[day][row['AwayTeam']] - goal_co[day][row['AwayTeam']])
        # On incrémente day quand on passe d'une journée à une autre
        if ((index + 1)% matches_per_day) == 0:
            day += 1
    
    df['MatchWeek'] = M_week
    df['HomeScored'] = H_scored
    df['HomeConceded'] = H_conceded
    df['AwayScored'] = A_scored
    df['AwayConceded'] = A_conceded
    df['HomeDiff'] = H_diff
    df['AwayDiff'] = A_diff

    return df
```

```python
def made_shot(df, covid):
    # Création du dictionnaire des équipes de la saison
    teams = {}
    for i in df['HomeTeam'].unique():
        teams[i] = []
    # Journée par journée, on ajoute le nombre de tirs réalisés
    for _, row in df.iterrows():
        shot_home = row['HS']
        shot_away = row['AS']
        teams[row['HomeTeam']].append(shot_home)
        teams[row['AwayTeam']].append(shot_away)
    # On ajoute ces données dans un dataframe équipe par équipe, journée par journée
    colonnes = list(range(1, 2*(len(teams)-1)+1))
    if covid:
        colonnes = list(range(1, 28))
    df_shot = pd.DataFrame(data = teams, index=colonnes).T
    df_shot.insert(0, 0, 0)
    # On cumule le nombre de tirs
    for i in colonnes:
        df_shot[i] = df_shot[i] + df_shot[i-1]
    return df_shot
```

```python
def conc_shot_target(df, covid):
    # Création du dictionnaire des équipes de la saison
    teams = {}
    for i in df['HomeTeam'].unique():
        teams[i] = []
    # Journée par journée, on ajoute le nombre de tirs cadrés concédés
    for _, row in df.iterrows():
        shot_home = row['HST']
        shot_away = row['AST']
        teams[row['HomeTeam']].append(shot_away)
        teams[row['AwayTeam']].append(shot_home)
    # On ajoute ces données dans un dataframe équipe par équipe, journée par journée
    colonnes = list(range(1, 2*(len(teams)-1)+1))
    if covid:
        colonnes = list(range(1, 28))
    df_shot = pd.DataFrame(data = teams, index=colonnes).T
    df_shot.insert(0, 0, 0)
    # On cumule le nombre de tirs cadrés
    for i in colonnes:
        df_shot[i] = df_shot[i] + df_shot[i-1]
    return df_shot
```

```python
def get_shots(df, covid):
    # On récupère les dataframes de tirs
    shot_ma = made_shot(df, covid)
    shot_co = conc_shot(df, covid)
    shot_ma_target = made_shot_target(df, covid)
    shot_co_target = conc_shot_target(df, covid)
    
    # On calcule le nombre de matches par journée
    matches_per_day = int(len(df)/(shot_ma.shape[1]-1))

    # On initialise les tableaux
    day = 0
    H_shot_ma = []
    H_shot_co = []
    H_shot_ma_target = []
    H_shot_co_target = []
    A_shot_ma = []
    A_shot_co = []
    A_shot_ma_target = []
    A_shot_co_target = []

    # Pour chaque match, on ajoute les tirs
    for index, row in df.iterrows():
        H_shot_ma.append(shot_ma[day][row['HomeTeam']])
        H_shot_co.append(shot_co[day][row['HomeTeam']])
        H_shot_ma_target.append(shot_ma_target[day][row['HomeTeam']])
        H_shot_co_target.append(shot_co_target[day][row['HomeTeam']])
        A_shot_ma.append(shot_ma[day][row['AwayTeam']])
        A_shot_co.append(shot_co[day][row['AwayTeam']])
        A_shot_ma_target.append(shot_ma_target[day][row['AwayTeam']])
        A_shot_co_target.append(shot_co_target[day][row['AwayTeam']])
        # On incrémente day quand on passe d'une journée à une autre
        if ((index + 1)% matches_per_day) == 0:
            day += 1
    
    df['HomeMadeShot'] = H_shot_ma
    df['HomeConcededShot'] = H_shot_co
    df['HomeMadeShotTarget'] = H_shot_ma_target
    df['HomeConcededShotTarget'] = H_shot_co_target
    df['AwayMadeShot'] = A_shot_ma
    df['AwayConcededShot'] = A_shot_co
    df['AwayMadeShotTarget'] = A_shot_ma_target
    df['AwayConcededShotTarget'] = A_shot_co_target

    return df
```

## Ajout des points

Création d'un dataframe de points des équipes par journée

```python
def result(df, covid):
    # Création du dictionnaire des équipes de la saison
    teams = {}
    for i in df['HomeTeam'].unique():
        teams[i] = []
    # Remplissage du ditionnaire (W:Victoire, L:Défaite, D:Match nul)
    for index, row in df.iterrows():
        if row['FTR'] == 'H':
            teams[row['HomeTeam']].append('W')
            teams[row['AwayTeam']].append('L')
        if row['FTR'] == 'D':
            teams[row['HomeTeam']].append('D')
            teams[row['AwayTeam']].append('D')
        if row['FTR'] == 'A':
            teams[row['HomeTeam']].append('L')
            teams[row['AwayTeam']].append('W')
    
    colonnes = list(range(1, 2*(len(teams)-1)+1))
    if covid:
        colonnes = list(range(1, 28))
    # Création du dataframe
    return pd.DataFrame(data=teams, index = colonnes).T
```

Fonction transformant un résultat textuel en points

```python
def points(string):
    if string == 'W':
        return 3
    if string == 'D':
        return 1
    if string == 'L':
        return 0
    else:
        return 0
```

Création d'un dataframe des points cumulés par journée pour chaque équipe

```python
def result_pts(results):
    # On transforme les résultats en points
    results = results.applymap(points)
    
    # On cumule le nombre de points
    for i in range(2,results.shape[1]+1):
        results[i] = results[i] + results[i-1]
    results.insert(0, 0, 0)

    return results
```

Ajout des points par journée au dataframe

```python
def get_points(df, covid):
    res = result(df, covid)
    res = result_pts(res)

    # On initialise les tableaux
    day = 0
    H_pts = []
    A_pts = []
    
    # On calcule le nombre de matches par journée
    matches_per_day = int(len(df)/(res.shape[1]-1))

    # Pour chaque match, on ajoute les points
    for index, row in df.iterrows():
        H_pts.append(res[day][row['HomeTeam']])
        A_pts.append(res[day][row['AwayTeam']])
        # On incrémente day quand on passe d'une journée à une autre
        if ((index + 1)% matches_per_day) == 0:
            day += 1
    
    df['HomePoints'] = H_pts
    df['AwayPoints'] = A_pts

    return df
```

## Ajout de la forme de l'équipe

Fontion renvoyant un tableau contenant les 'num' résultats récents de chaque équipe par journée

```python
def form(df, num, covid):
    res = result(df, covid)
    new_res = res.copy()
    # On ajoute les résultats récents
    for i in range(num, res.shape[1]+1):
        new_res[i] = ''
        j = 0
        while j < num:
            new_res[i] += res[i-j]
            j += 1
    return new_res
```

Ajout des `num` derniers résultats récents par journée au dataframe

```python
def num_form(df, num, covid):
    forme = form(df, num, covid)
    # Création de listes de séries
    # Les premiers matchs, ou on ne peut pas avoir de série, sont marqués 'X'
    home = ['X' for i in range(num*int(len(forme)/2))]
    away = ['X' for i in range(num*int(len(forme)/2))]

    matches_per_day = int(len(df)/(forme.shape[1]-1))

    day = num
    #for i in range(num*int(len(forme)/2), len(df)):
    for index, row in df.iterrows():
        if index >= num*int(len(forme)/2):
            home_form = forme[day][row['HomeTeam']]
            away_form = forme[day][row['AwayTeam']]
            # On ajoute le 'num' dernier résultat
            home.append(home_form[num-1])
            away.append(away_form[num-1])
            # On incrémente day quand on passe d'une journée à une autre
            if ((index + 1)% matches_per_day) == 0:
                day += 1
    
    # On ajoute la série au dataframe
    df['HS'+str(num)] = home
    df['AS'+str(num)] = away

    return df
```

Ajout des derniers résultats récents par journée au dataframe

```python
def get_form(df, covid):
    for i in range(1,6):
        df = num_form(df, i, covid)
    return df
```
